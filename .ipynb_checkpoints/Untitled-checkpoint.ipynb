```python
df["type"].value_counts().plot (kind="bar", title="Transaction Type", color="blue")
plt.xlabel("Transaction Type")
plt.ylabel("Count")
plt.show()
```

```python
fraud_by_type=df.groupby("type")["isFraud"].mean().sort_values(ascending=False)
fraud_by_type.plot(kind="bar",title="fraud rate by type",color="green")
plt.xlabel("fraud rate")
plt.show()
```

```python
df["amount"].describe().astype(int)
```

```python
sns.histplot(np.log1p(df["amount"]),bins=100,kde=True,color="red")
plt.title("transaction amount distribution log scale")
plt.xlabel("log(amount+1)")
plt.show()
```

```python
sns.boxplot(data=df[df["amount"]<50000],x="isFraud",y="amount")
plt.title("amount vs isfraud(filter under 50k)")
plt.show()
```

```python
df["balanceDiffOrig"] = df["oldbalanceOrg"] - df["newbalanceOrig"]
df["balanceDiffDest"] = df["newbalanceDest"] - df["oldbalanceDest"]
```

```python
frauds_per_step=df[df["isFraud"] == 1]["step"].value_counts().sort_index()
plt.plot(frauds_per_step.index , frauds_per_step.values, label="frauds per step") 
plt.xlabel("step(time)")
plt.ylabel=("no of frauds")
plt.title("frauds over time")
plt.grid(True)
plt.show()
```
